```python
# prompt: apply label encoder on object columns and save them in a list of pickles

import pickle # Import the pickle library for saving and loading objects
from sklearn.preprocessing import LabelEncoder # Import LabelEncoder for categorical feature encoding

# Identify object columns (columns with 'object' data type)
object_cols = df.select_dtypes(include='object').columns

# Create a list to store encoders
encoders = []

# Apply Label Encoding to object columns and save encoders
for col in object_cols:
    le = LabelEncoder() # Initialize a LabelEncoder for each column
    df[col] = le.fit_transform(df[col]) # Fit and transform the column, replacing original values with encoded labels
    encoders.append({col: le}) # Store encoder with column name in the list

# Save the list of encoders as a pickle file to Google Drive
with open('/content/drive/MyDrive/Main_Project/label_encoders.pkl', 'wb') as f:
    pickle.dump(encoders, f)

# Display info after encoding
df.info()
df.head()
```

```python
%%writefile app.py
import pickle
import pandas as pd
import numpy as np
import streamlit as st

# Load trained model and label encoder from saved files
model1 = pickle.load(open("model.pkl", "rb"))
label_encoder = pickle.load(open("/content/drive/MyDrive/Main_Project/label_encoders.pkl", "rb"))

# Function to preprocess input data using the loaded label encoders
def preprocess(df):
    # Label encoder was saved as a list of dictionaries, each containing {column_name: encoder_object}
    # We need to iterate through the list and apply the correct encoder to each column.
    for encoder_dict in label_encoder:
        for col_name, le in encoder_dict.items():
            try:
                df[col_name] = le.transform(df[col_name])
            except ValueError as e:
                 if "y contains previously unseen labels" in str(e):
                    # Handle unseen labels by adding them to the encoder's classes
                    unseen_label = str(e).split(': ')[-1].strip()[1:-1]
                    le.classes_ = np.append(le.classes_, unseen_label)
                    df[col_name] = le.transform(df[col_name])
                 else:
                    raise e
    return df

# Function to make predictions using the loaded model
def predict(input_data):
    # Ensure input data has the same features and order as the training data
    training_features = model1.feature_names_in_
    input_data = input_data.reindex(columns=training_features, fill_value=0)
    prediction = model1.predict(input_data) # Make prediction
    return prediction[0] # Return the predicted value

# Streamlit UI to get user input for prediction
st.title("Real Estate Price Prediction") # Set the title of the Streamlit app

# Input fields for user to enter property details
bedrooms = st.number_input("Number of Bedrooms", min_value=1, value=1)
baths = st.number_input("Number of Baths", min_value=1, value=1)
total_area = st.number_input("Total Area (m)", min_value=0.0, value=100.0)

# Updated predefined categories for dropdown menus
property_type_options = ['Flat', 'House', 'Penthouse', 'Farm House', 'Lower Portion', 'Upper Portion', 'Room']
property_type = st.selectbox("Property Type", property_type_options, index=0)

city_options = ['Islamabad', 'Lahore', 'Faisalabad', 'Rawalpindi', 'Karachi']
city = st.selectbox("City", city_options, index=0)

province_options = ['Islamabad Capital', 'Punjab', 'Sindh']
province = st.selectbox("Province", province_options, index=0)

location = st.text_input("Location", value="")

latitude = st.number_input("Latitude", min_value=-90.0, max_value=90.0, value=0.0)
longitude = st.number_input("Longitude", min_value=-180.0, max_value=180.0, value=0.0)

purpose_options = ['For Sale', 'For Rent']
purpose = st.selectbox("Purpose", purpose_options, index=0)

# Predict price button
if st.button("Predict Price"):
    # Create a DataFrame from user input
    df = pd.DataFrame([[bedrooms, baths, total_area, property_type, city, province, location, latitude, longitude, purpose]],
                      columns=['bedrooms', 'baths', 'Total_Area', 'property_type', 'city', 'province_name', 'location', 'latitude', 'longitude', 'purpose'])
    input_data = preprocess(df) # Preprocess the input data
    predicted_price = predict(input_data) # Get the predicted price (scaled)
    xmin = 0 # Minimum price from training data (for inverse scaling)
    xmax = 2000000000 # Maximum price from training data (for inverse scaling)
    original_price = predicted_price * (xmax - xmin) + xmin # Inverse scale the predicted price to original range
    st.write(f"Predicted Price: {original_price}") # Display the predicted original price
```
